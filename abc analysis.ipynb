Count sales grouped by products in descending order

```python
grouped_df = df.groupby('DR_NDrugs').agg({'DR_Kol':sum})
```

```python
## proportion of sales of each product grouped by products and sorted in descending order
grouped_df = (grouped_df['DR_Kol']/sum(grouped_df['DR_Kol'])).sort_values(ascending=False)
```

Count sum of sales in cumulative total (накопительным итогом)

```python
grouped_df = grouped_df.cumsum().reset_index()
grouped_df
```

Define group of product. In abc-analysis there are 3 groups: A, B, C based on Pareto rule. Group A products give 80% of sales, group B gives 15% of sales (80-95%), group C products give 5% of sales (95-100%) ordered in descending order in cumulative total. 

```python
import numpy as np
```

```python
grouped_df['abc']=np.where(grouped_df['DR_Kol'] < 0.8, 'A', np.where(grouped_df['DR_Kol'] < 0.95, 'B', 'C'))
```

Let's write a universal function not to copypaste code

```python
def perform_abc(df, index):
    """
    на вход подает датафрейм пандас...
    """
    cols = list(df.columns)
    cols.remove(index)
    grouped_df = df.groupby(index).agg({col: sum for col in cols})
    for col in cols:
        grouped_df[f'rel_{col}'] = grouped_df[col]/sum(grouped_df[col])
        grouped_df=grouped_df.sort_values(f'rel_{col}', ascending=False)
        grouped_df[f'cumsum_{col}'] = grouped_df[f'rel_{col}'].cumsum()
        grouped_df[f'abc_{col}']=np.where(grouped_df[f'cumsum_{col}'] < 0.8, 'A', np.where(grouped_df[f'cumsum_{col}'] < 0.95, 'B', 'C'))
    return grouped_df[cols + [f'abc_{col}' for col in cols]]
```

```python
df1 = df[['DR_NDrugs', 'DR_CRoz', 'DR_Kol']]
```

```python
perform_abc(df1, index='DR_NDrugs')
```

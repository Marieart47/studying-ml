```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Izumr.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.set_index('names')
```

```python
df.plot(kind = 'scatter', x = 'X', y = 'Y')
plt.minorticks_on()
plt.grid(which='minor', color='#aaa', ls=':')
plt.show()
```

```python
plt.figure(1, figsize = (16, 8))
tree_plot_complete = dendrogram(linkage(df, method  = "complete", metric="euclidean"))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
plt.figure(1, figsize = (16, 8))
tree_plot_average = dendrogram(linkage(df, method  = "average", metric="euclidean"))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
#  
plt.figure(1, figsize = (16, 8))
tree_plot_single = dendrogram(linkage(df, method  = "single", metric="euclidean"))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Euclidean distances')
plt.show()
```

Берут помесячные (поквартальные) объемы продаж по разным продуктам за какой-то длительный период (не менее 4 периодов), далее считается коэффицент вариации (стандоткл делить на срзнач), на основании коэффициента вариации относим к группе: X - маленький коэффициент вариации (<10%) - это товра с устойчивым спросом, не нужно иметь большой запас на складе, достаточно часто закупать примерно одинаковое количество, Y - не очень стабильные продажи (10-25%), спрос можно примерно спрогнозировать, на складе нужно иметь запас но не перезатарку, Z - очень сезонный товар, неконтролируемый спрос, нужно затарить склад этим товаром (>25%)
